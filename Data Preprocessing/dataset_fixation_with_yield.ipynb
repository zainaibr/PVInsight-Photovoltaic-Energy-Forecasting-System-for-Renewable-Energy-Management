```python
sensor_cols = [
    "irr_sensor_1", "irr_sensor_2", "irr_sensor_3",
    "irr_sensor_4", "irr_sensor_5", "irr_sensor_6",
    "irradiance_avg"
]

def count_sensor_outliers(df, columns):
    outlier_counts = {}
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        outlier_counts[col] = df[(df[col] < lower) | (df[col] > upper)].shape[0]
    return outlier_counts

outliers = count_sensor_outliers(df, sensor_cols)
print("Number of outliers per sensor:\n", outliers)
```

```python
import pandas as pd
import glob
import os


files = glob.glob("*.xlsx")
output_file = "trial_display_file.xlsx"

month_order = {
    "january": 1,
    "february": 2,
    "march": 3,
    "april": 4,
    "may": 5,
    "june": 6,
    "july": 7,
    "august": 8,
    "september": 9,
    "october": 10,
    "november": 11,
    "december": 12
}

files = sorted(
    files,
    key=lambda f: next(
        (month_order[m] for m in month_order if m in os.path.basename(f).lower()),
        99
    )
)

print("Files found:")
for f in files:
    print("-", os.path.basename(f))

df_merged = pd.DataFrame()

for file in files:
    filename = os.path.basename(file)
    print(f"\nReading: {filename}")

    df = pd.read_excel(file, skiprows=3)

    df.columns = [
    "datetime",
    "col_unused",
    "irr_sensor_1",
    "irr_sensor_2",
    "irr_sensor_3",
    "irr_sensor_4",
    "irr_sensor_5",
    "irr_sensor_6",
    "irradiation_tilted",
    "power_analyzer",
    "generated_yield",
    "avg_module_temp",
    "module_temp",
    "1",
    "2",
    "30",
    "4",
    "5",
    "6",
    "7",
    "8",
    "9"
]

    df = df[[
        "datetime",
        "irr_sensor_1",
        "irr_sensor_2",
        "irr_sensor_3",
        "irr_sensor_4",
        "irr_sensor_5",
        "irr_sensor_6",
        "irradiation_tilted",
        "power_analyzer",
        "generated_yield",
        "avg_module_temp",
    ]]

    df["datetime"] = pd.to_datetime(
        df["datetime"],
        format="%d.%m.%Y %H:%M:%S",
        errors="coerce"
    )

    df = df.dropna(subset=["datetime"])
    df = df.set_index("datetime")

    # Convert relevant columns to numeric, coercing errors to NaN
    numeric_cols = [
        "irr_sensor_1", "irr_sensor_2", "irr_sensor_3",
        "irr_sensor_4", "irr_sensor_5", "irr_sensor_6",
        "irradiation_tilted", "power_analyzer", "generated_yield",
        "avg_module_temp"
    ]
    for col in numeric_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce')

    # Calculate the average irradiance for the raw data
    df["irradiance_avg"] = df[[
        "irr_sensor_1", "irr_sensor_2", "irr_sensor_3",
        "irr_sensor_4", "irr_sensor_5", "irr_sensor_6"
    ]].mean(axis=1, skipna=True)

    # Resample to hourly mean after ensuring all columns are numeric
    df = df.resample("15T").mean()

    # Filter out hours where the overall mean irradiance is below a threshold
    # This line depends on 'irradiance_avg' being present and numeric.
    # If 'irradiance_avg' is not critical for the filter, consider using another column or the mean of other columns.
    #df = df[df["irradiance_avg"] > 50] ## remv

    df["source_file"] = filename

    df_merged = pd.concat([df_merged, df], axis=0)

    print("Current merged shape:", df_merged.shape)

df_merged = df_merged.sort_index()
df_merged.to_excel(output_file)

print("\nFinal merged shape:", df_merged.shape)
print("Saved to:", output_file)

display(df_merged.head())
display(df_merged.tail())
```

```python
AREA = 60000  # m²
gamma = -0.004  # temperature coefficient (per °C)
T_ref = 25  # reference temperature
```

# Convert energy → power

# The generated yield represents the *energy output over each 15-minute interval*, not instantaneous power. Therefore, I convert it to power by dividing by the time interval (0.25 hours) before applying physical equations.

```python
# 15 minutes = 0.25 hour
df["power_kw"] = df["generated_yield"] / 0.25
```

```python
df.head(50)
```

#  FIX UNITS (VERY IMPORTANT)

irradiance is in W/m², and power is in kW

convert irradiance to kW/m²

```python
df["irradiance_kw_m2"] = df["irradiance_avg"] / 1000
```

# Compute efficiency correctly 𝜂=𝑃/𝐺×𝐴

```python

df["efficiency"] = df["power_kw"] / (
    df["irradiance_kw_m2"] * AREA
)
```

```python
df.head(50)
```

# The 50 W/m² threshold is used to remove unstable efficiency values, , The usage of values such as 20 or so will add low irradiance values to  that denominator that explode the values of the efficiency and affect the reference efficiency

```python
df.loc[df["irradiance_avg"] < 50, "efficiency"] = pd.NA
df["efficiency"] = df["efficiency"].clip(lower=0, upper=1)
```

# 200 W/m² threshold is used to ensure that reference efficiency is calculated only from high-quality, stable operating conditions.Since the basline effieciency should be presented with the best values as it is the bottle neck of optiminmzing the system and conducting analysis.

```python
eta_ref = df[df["irradiance_avg"] > 200]["efficiency"].mean()
print("η_ref =", eta_ref)
```
